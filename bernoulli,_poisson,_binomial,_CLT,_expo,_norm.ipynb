# Bernoulli distribution

```python
from scipy.stats import bernoulli
```

```python
# p=0.5 i.e a fair coin probability
bernoulli.rvs(p=0.5, size = 10) # size specify = number of trials
# rvs = random variates - to generate random numbers with probability p defined
#bernoulli can be seen as binomial random variable with n=1.
```

```python
bernoulli.rvs(p=0.2, size = 10) #negatively biased coin
```

```python
bernoulli.rvs(p=0.8, size=10) #positively biased coin
```

```python
#plotting the distribution for N numbers of trials
N_trials = [10,25,100,250,1000,10000,25000] #number of trials
pr = 0.5 # fair coin toss probablity
av = [] #empty list to store the average of random variates
```

```python
#Generate 10 variates everytime and take the avergae. That should be of 1's i.e 0.5 for fair coin
for i in N_trials:
      for n in range(1,i+1):
          av.append(np.mean(bernoulli.rvs(p=pr, size=10))) #append adds element to the end of list
#to calculate mean called np = numpy and added that element into av[] = empty list
      if(i==10):
        plt.title("distribution with {} trials of 10 coin tosses".format(i))
        plt.hist(av,bins=10) # bins define equal no. of spaces or width in the range.
        plt.xlim(0.0,1.0)
        plt.show()
      else:
          plt.title("distribution with {} trials of 10 coin tosses".format(i))
          plt.hist(av,bins=25)
          plt.xlim(0.0,1.0)
          plt.show()
```

```python
print(av)
```

```python
print("A fair coin is spinning...\n"+"-"*30)
pr=0.5 #fair coin toss probability
# mvsk are the four moments in bernoulli
mean, var = bernoulli.stats(p=pr, moments = 'mv') #mean, var, skew, kurt = bernoulli.stats(p, moments='mvsk')
print("mean:", mean)
print("Variance:", var)
#The mean of a Bernoulli distribution is E[X] = p and the variance, Var[X] = p(1-p).
#bernoulli is special case of binomial distribution when only one trial is conducted.
```

```python
from scipy.stats import binom
#this binomial random varibale counts the number of successes in n independent trials of random experiment where the probability of success is p.
```

# Poisson Distribution
have 2 parameters
n --> infinity
p --> 0

```python
from scipy.stats import poisson
#this random variable counts the no. of successes in n independent bernoulli trials
#here the parameter which is used to shape the distribution is lambda or mu = np >=0
#where n= no. of trials and p = probability of success n--> infinity, p--> 0 
#instead of n, p like in above two, here we just use mu.
```

```python
la = 5
r= poisson.rvs(mu=la, size=20)
print("random variates with lambda={}:{}".format(la,r))

la=0.5
r= poisson.rvs(mu=la, size=20)
print("random variates with lambda{}:{}".format(la,r))
```

```python
#to generate random poisson variates
mu=2
rv=poisson.rvs(mu,size=10)
print(rv)
```

```python
la=0.5
rv=poisson(la)
x=np.arange(0,21,1)
pmf1=rv.pmf(x)

la=1
rv=poisson(la)
x=np.arange(0,21,1)
pmf2=rv.pmf(x)

la=5
rv=poisson(la)
x=np.arange(0,21,1)
pmf3= rv.pmf(x)

plt.figure(figsize=(9,6))
plt.title("Probability mass function: $e^{-\lambda}{(\lambda^k/k!)}$\n",fontsize=20) # https://matplotlib.org/stable/tutorials/text/mathtext.html
#refer link for learning about mathematical expression
plt.scatter(x,pmf1,s=100) # s gives size of the dot on the scatter plot
plt.scatter(x,pmf2,s=100)
plt.scatter(x,pmf3,s=100)
plt.legend(["$\lambda=0.5$","$\lambda=1$","$\lambda=5$"],fontsize=15)
plt.xlabel("Number of occurences($k$",fontsize=15)
plt.ylabel("$Pr(X=k)$",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.show()
```

```python
la=0.5
rv=poisson(la)
x=np.arange(0,21,1)
cdf1= rv.cdf(x)

la=2
rv=poisson(la)
x=np.arange(0,21,1)
cdf2=rv.cdf(x)

la=5
rv=poisson(la)
x=np.arange(0,21,1)
cdf3=rv.cdf(x)

plt.figure(figsize=(9,6))
plt.title("cumulative distribution function\n",fontsize=20)
plt.scatter(x,cdf1,s=100)
plt.scatter(x,cdf2,s=100)
plt.scatter(x,cdf3,s=100)
plt.legend(["$\lambda=0.5$","$\lambda=2$","$\lambda=5$"],fontsize=15)
plt.xlabel("number of occurences($k$)",fontsize=15)
plt.ylabel("cumulative distribution function",fontsize=15)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.grid(True)
plt.show()
```

```python
fig,ax=plt.subplots(1,1)
r=norm.rvs(size=1000)
x=np.linspace(norm.ppf(0.01),norm.ppf(0.50),100)
#ax.plot(x,norm.pdf(x),'r-',lw=1,alpha=0.6,label='norm pdf') #lw=linewidth
rv=norm()
ax.plot(x, rv.pdf(x),'k-',lw=2,label='frozen pdf')
ax.hist(r,density=True, histtype='stepfilled',alpha=0.2) #stepfilled generates a plot that is by default filled
ax.legend(loc='best',frameon=False)
plt.show()
```

# Exponential distribution

```python
from scipy.stats import expon
```

```python
fig,ax=plt.subplots(1,1)
x=np.linspace(expon.ppf(0.01),expon.ppf(0.99),100)
ax.plot(x,expon.pdf(x),'r-',lw=5,alpha=0.6,label='expon pdf')
rv=expon()
ax.plot(x,rv.pdf(x),'k-',lw=2, label='frozen pdf')
r=expon.rvs(size=1000)
ax.hist(r,density=True,histtype='stepfilled',alpha=0.2)
ax.legend(loc='best',frameon=False)
plt.show()
```

# central limit theorem

```python
import seaborn as sbn
```

```python
from scipy.stats import uniform #import unform continuous random variable
#distribution is uniform on [0,1]
```

```python
sample_data=np.random.normal(size=100000)
plt.hist(sample_data)
plt.show() #to plot histogram without axises
sbn.distplot(sample_data)
sbn.distplot(sample_data,kde=True) #kde= kernel density estimate = way to estimate pdf of continuous random variable
# kde= true will show normal distribution line and kde=false will remove it.
```

```python
def apply_CLT(sample_data,sample_size, total_samples):
  sample_mean=[]
  for a in range(total_samples): #iterating for total no. of samples for choosing samples and their mean
    sample = np.random.choice(sample_data, size=sample_size,replace=True) #generates a random sample for given 1-D array
    # here repleace = false means without replacement - cant select sample_data multiple times
    mean=np.mean(sample)
    sample_mean.append(mean)
  return sample_mean
  #distribution of sample tends towards normal distribution as size of sample increases.
```

```python
#applying CLT on sample data. Function = apply_CLT(sample_data,sample_size, total_samples)
sbn.distplot(apply_CLT(sample_data,30,1000))
```

```python
sample_data=uniform.rvs(size=1000)
sbn.distplot(sample_data)
```

```python
sbn.distplot(apply_CLT(sample_data,10000,10000))
# realisation- as sample size increases, we can observe that it takes shape of normal distribution.
```
